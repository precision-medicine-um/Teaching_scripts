```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.zeros([100, 100])
```

```python
plt.imshow(A)
plt.show()
```

```python
A.shape
```

```python
x_center = int(A.shape[0]/2)
y_center = int(A.shape[1]/2)
```

```python
R = 25
```

```python
import math
```

```python
for x in range (0, A.shape[0]):
    for y in range (0, A.shape[1]):
        if math.sqrt((x-x_center)**2+(y-y_center)**2)<R:
            A[x,y] = 1
```

```python
plt.imshow(A, vmin=0, vmax=1)
plt.show()
```

```python
from google.colab import files
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img = cv2.imread(filename)
img_3ch = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_3ch) # Display
plt.axis('off')
plt.show()

print(f"Image size {img_3ch.shape}")
```

```python
img_3ch.shape
```

```python
plt.imshow(img_3ch)
plt.show()
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(131)
plt.imshow(img_3ch[..., 0])

plt.subplot(132)
plt.imshow(img_3ch[..., 1])

plt.subplot(133)
plt.imshow(img_3ch[..., 2])

plt.show()
```

```python
img_bw = np.mean(img_3ch, axis = 2)
```

```python
plt.imshow(img_bw)
plt.show()
```

```python
plt.hist(img_bw.flatten(), bins = 100)
plt.show()
```
